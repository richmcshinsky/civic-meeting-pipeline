Quick look at the ingested canonical records before dedup and extraction. Run `make ingest` first, then this notebook reads `data/interim/canonical.jsonl`.

```python
import json
from pathlib import Path
import pandas as pd

path = Path('../data/interim/canonical.jsonl')
rows = [json.loads(l) for l in path.open()] if path.exists() else []
df = pd.DataFrame(rows)
print(len(df), 'records')
df.head()
```

```python
# Coverage by source and jurisdiction
if len(df):
    display(df.groupby(['source', 'jurisdiction_name']).size().rename('n'))
```
